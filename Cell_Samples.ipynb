```python
import pandas as pd
```

```python
df = pd.read_csv('Data/cell_samples.csv')
```

```python
df
```

```python
# describe the features in our data

Features = ['Clump','UnifSize','UnifShape', 'MargAdh',
'SingEpiSize', 'BareNuc' ,'BlandChrom', 'NormNucl', 'Mit', 'Class']

Description = ['Clump Size', 'Uniform Size', 'Uniform Shape', 'Marginal Adhesion',
               'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 
               'Normal Nucleoli', 'Mitoses', 'Benign or Malignant']

data = list(zip(Features,Description))

features_index = pd.DataFrame(data, columns=(['Features', 'Description']))
  
features_index
```

```python
#create a sql database 

import sqlite3 
conn = sqlite3.connect('Cell_db')
c = conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS description (Features text,Description text)')
conn.commit()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS features (Clump int,UnifSize int,UnifShape int, MargAdh int, SingEpiSize int, BareNuc int, BlandChrom int, NormNucl int, Mit int, Class int)')
conn.commit()
```

```python
features_index.to_sql('description', conn, if_exists='replace', index=False)
df.to_sql('features', conn, if_exists='replace', index=False)
```

```python
#call sql database to check if correct

c.execute('''
select *
from description ''')
new_df = pd.DataFrame(c.fetchall(), columns = ['Features', 'Description'])

new_df
```

```python
#call sql database to check if correct

c.execute('''
select *
from features ''')
new_df2 = pd.DataFrame(c.fetchall(), columns = ['Clump','UnifSize','UnifShape', 'MargAdh',
'SingEpiSize', 'BareNuc' ,'BlandChrom', 'NormNucl', 'Mit', 'Class'])

new_df2
```
